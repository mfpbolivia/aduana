```python
##Importamos Libreria
import requests
from bs4 import BeautifulSoup
import pandas as pd

#guardamos en variable la url de donde obtendremos los datos
baseurl="http://anbsw04.aduana.gob.bo:7551/subastas/lotelista.do?parameter=catalogo"

```

```python
#Realizamos un get a a url
r=requests.get(baseurl)
#guardamos en soup el contenido 
soup=BeautifulSoup(r.content,'html.parser')
#Obtenemos en listado todos los tags tr encontrados
listado=soup.find_all('tr',class_='habCat')

```

```python
#Creamos un dataframe vacio
df=pd.DataFrame()
df['id']=None
df['lugar']=None
df['articulo']=None
df['almacen']=None
df['direccion']=None
df['fecha']=None
df['estado']=None
```

```python
#creamos unq lista vacias para anadir las filas que encntremos en listado
datos=[]
```

```python
#Recorremos listado y guardamos los resultados de cada tag td
for item in listado:
    dat=item.find_all('td')
    id=dat[0].text
    lugar=dat[1].text
    articulo=dat[2].text
    almacen=dat[3].text
    direccion=dat[4].text
    fecha=dat[5].text
    estado=dat[6].text
    #hacemos un append a la lista datos de cada fila encontrada
    datos.append([id,lugar,articulo,almacen,direccion,fecha,estado])
```

```python
#Subimos la lista datos al dataframe
df = pd.DataFrame(datos)
```

```python
#Renombramos las columnas
df.rename(columns={0:'id',1:'lugar',2:'articulo',3:'almacen',4:'direccion',5:'fecha',6:'estado'},inplace=True)
```

```python
#cambiamos el tipo de dato de object a fecha a la columna con nombre=fecha
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
#Eliminamos espacios saltos de linea y otros de la columna estado
df['estado']=df['estado'].str.strip()
```
