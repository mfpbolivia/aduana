```python
import requests
import pdfplumber
import pandas as pd
import re
```

```python
def download_file(url):
  local_filename=url.split('/')[-1]

  with requests.get(url) as r:
    with open(local_filename, 'wb') as f:
      f.write(r.content)

  return local_filename
```

```python
ap_url='sla1.pdf'
```

```python
#ap=download_file(ap_url)
local_filename=ap_url.split('/')[-1]
```

```python
local_filename
```

```python
with pdfplumber.open(local_filename) as pdf:
  page=pdf.pages[0]
  text=page.extract_text()

print(text)
```

```python
titulo1=re.compile(r'Especialidades*')
titulo2=re.compile(r'Contacto*')
titulo3=re.compile(r'Telefono*')
```
